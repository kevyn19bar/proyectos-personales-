2. buscar_contacto(agenda, nombre): Devuelve la información de un contacto. Si no existe, debe indicarlo.

3. mostrar_contactos(agenda): Muestra todos los contactos de la agenda.

```python
def menu():
  print("Menu de agenda de contactos")
  print("1. _Agregar nuevo contacto_")
  print("2. ____Buscar contacto_____")
  print("3. _____Mostrar agenda_____")
  print("4. _________Salir__________")
def agregar_contacto():
  nombre = input("Ingrese el nombre del contacto: ")
  info={"telefono":input("Ingrese el numero de telefono: "),"email":input("Ingrese el email: ")}
  return nombre, info
def buscar_contacto(agenda, buscar):
  mensaje=""
  for nombre in agenda:
    if buscar == nombre:
      mensaje="informacion de ", buscar , " es ", agenda[nombre]
    else:
      mensaje="no existe informacion de "+ buscar
  return mensaje
def mostrar_contactos(agenda):
  print(agenda)
continuar=True
agenda={}
while continuar:
  menu()
  op=int(input("Elige una opcion: "))
  if op==1:
    clave,valor=agregar_contacto()
    agenda[clave]=valor
  elif op==2:
    nombre=input("Ingrese el nombre de conctacto que quieres buscar: ")
    respuesta=buscar_contacto(agenda, nombre)
    print(respuesta)
  elif op==3:
    mostrar_contactos(agenda)
  elif op==4:
    continuar=False
  else:
    print("Opcion invalidad \n vuelva a intentarlo")
```

#**Ejercicio 4: Tablero de Tres en Raya**
Conceptos: Funciones, Listas Dobles (listas anidadas).

Enunciado:

Representa un tablero de "Tres en Raya" (Tic-Tac-Toe) usando una lista de listas de 3x3.

1. Crea una función crear_tablero()que retorna una matriz de 3x3 con espacios en blanco ( ' ').

2. Escribe una función imprimir_tablero(tablero)que muestre el tablero en la consola de forma clara.

3. Implementa una función hacer_movimiento(tablero, fila, columna, jugador)que coloca la ficha del jugador ( 'X'o 'O') en la posición indicada.

```python
def crear_tablero():
    return [[' ' for _ in range(3)] for _ in range(3)]

def imprimir_tablero(tablero):
    print("\n-------------")
    for fila in tablero:
        print(f"| {fila[0]} | {fila[1]} | {fila[2]} |")
        print("-------------")

def verificar_ganador(tablero, jugador):
    for i in range(3):
        if all(tablero[i][j] == jugador for j in range(3)) or \
           all(tablero[j][i] == jugador for j in range(3)):
            return True
    if all(tablero[i][i] == jugador for i in range(3)) or \
       all(tablero[i][2 - i] == jugador for i in range(3)):
        return True
    return False

def tablero_lleno(tablero):
    for fila in tablero:
        if ' ' in fila:
            return False
    return True

def jugar_gato():
    tablero = crear_tablero()
    jugador_actual = 'X'

    print("Bienvenido al tres en raya")

    while True:
        imprimir_tablero(tablero)
        print(f"Turno del jugador: {jugador_actual}")

        try:
            fila = int(input("Elige una fila (0, 1, 2): "))
            columna = int(input("Elige una columna (0, 1, 2): "))

            if fila in range(3) and columna in range(3):
                if tablero[fila][columna] == ' ':
                    tablero[fila][columna] = jugador_actual
                else:
                    print("Esa casilla ya está ocupada Inténtalo de nuevo.")
                    continue
            else:
                print("Posición inválida Elige números entre 0 y 2.")
                continue

        except ValueError:
            print("Entrada inválida Por favor, introduce solo números.")
            continue

        if verificar_ganador(tablero, jugador_actual):
            imprimir_tablero(tablero)
            print(f"Felicidades El jugador {jugador_actual} ha ganado.")
            break

        if tablero_lleno(tablero):
            imprimir_tablero(tablero)
            print("El juego es un empate")
            break

        jugador_actual = 'O' if jugador_actual == 'X' else 'X'

jugar_gato()
```

#**Ejercicio 6: Suma de Elementos en una Matriz**
Conceptos: Funciones, Listas Dobles.

Enunciado:

Escribe una función sumar_matriz(matriz)que recibe una matriz (lista de listas con números) y calcula la suma total de todos sus elementos.

```python
def sumar_matriz(matriz):
  suma=0
  for x in matriz:
    suma+=sum(x)
  return suma
matriz = [[1, 2, 3],[4, 5, 6],[7, 8, 9]]
resultado=sumar_matriz(matriz)
print("La suma total de la matriz es",resultado)
```

#**Ejercicio 7: Suma por Filas en una Matriz**
Conceptos: Funciones, Listas Dobles.

Enunciado:

Escribe una función suma_por_filas(matriz)que reciba una matriz y devuelva una nueva lista donde cada elemento sea la suma de los elementos de la fila correspondiente en la matriz original.

```python
def suma_por_filas(matriz):
  nueva_matriz=[]
  for x in matriz:
    suma=sum(x)
    nueva_matriz.append(suma)
  return nueva_matriz
matriz = [[10, 5, 3],[2, 20, 1],[7, 8, 15]]
resultado=suma_por_filas(matriz)
for z in range(len(resultado)):
  print("La suma de la fila",z+1,"es",resultado[z])
```
