#****Ejercicio N1****

Escribe un programa que reciba un numero entero y calcule la suma de todos los numeros pares desde 1 hasta ese numero utilizando ciclo

```python
num=int(input("Ingrese un numero entero: "))
suma=0
for x in range(1,num+1):
  if x%2==0:
    suma+=x
print("La suma de los numeros pares desde 1 hasta",num,"es:",suma)
```

#****Ejercicio N2****

programa que reciba un numero entero como limite y, dentro de un ciclo, calcule cuantas veces se repiten las vocales (a,e,i,o,u) entre 1 y ese numero

```python
!pip install num2words
from num2words import num2words
limite=int(input("Ingrese el limite: "))
Cont_a=0
Cont_e=0
Cont_i=0
Cont_o=0
Cont_u=0
for x in range(1,limite+1):
  numero=num2words(x,lang="es")
  for y in numero:
    if y=="a" or y=="á":
      Cont_a+=1
    elif y=="e" or y=="é":
      Cont_e+=1
    elif y=="i" or y=="í":
      Cont_i+=1
    elif y=="o" or y=="ó":
      Cont_o+=1
    elif y=="u" or y=="ú":
      Cont_u+=1
print("La cantidad de veces que se repite la vocal a es: ",Cont_a)
print("La cantidad de veces que se repite la vocal e es: ",Cont_e)
print("La cantidad de veces que se repite la vocal i es: ",Cont_i)
print("La cantidad de veces que se repite la vocal o es: ",Cont_o)
print("La cantidad de veces que se repite la vocal u es: ",Cont_u)
```

#****Ejercicio N5****

Crea un programa que reciba un numero entero y determine si los numeros entre 2 y ese numero son primos, sin usar lista. Utiliza ciclos para verificar si un numero es primo

```python
num=int(input("Ingrese un numero: "))
print("Los numeros primos del",num,"son:")
while num>1:
  primo=0
  for x in range(1,num+1):
    if num % x ==0:
      primo+=1
  if primo==2:
    print(num, end=" ")
  num-=1
```

#****Ejercicio N6****

Escribe un programa que reciba una cantidad de numeros y, mediante un ciclo, cuente cuantos de ellos son negativos, sin usar lista

#****Ejercicio N7****

Crea un programa que reciba una cantidad N de numeros (uno por uno) y calcule el promedio de esos numeros utilizando ciclo. No utilices lista, solo variables para almacenar la suma y el contador

```python
cand=int(input("Ingrese la cantidad de numeros: "))
Cont=0
suma=0
for x in range(cand):
  numero=int(input("Ingrese el numero: "))
  suma+=numero
  Cont+=1
promedio=suma/Cont
print("El promedio de los numeros es",promedio)
```

#****Ejercicio N9****

Crea un prgrama que invierta los digitos de un numero sin usal lista. Puede hacer esto utilizando un ciclo while y operaciones matematicas

```python
num=int(input("Ingrese un numero: "))
invertido=0
while num>0:
  dig=num%10
  invertido=invertido*10+dig
  num=num//10
print("El numero invertido es:",invertido)
```

#****Ejercicio N10****

Escribe un programa que reciba una cantidad N de numeros y determine cual es el mayor de todos sin usar estructuras de datos. Solo utiliza una variable para almacenar el valor mas alto encontrado hasta el momento
