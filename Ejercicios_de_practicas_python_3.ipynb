#****Ejercicio de cadena N1****

Escribe una funcion que reciba una cadena y devuelva cuantas vocales contiene.

```python
def contadorvucal(cadena):
  vocales="aeiouAEIOUáéíóúÁÉÍÓÚ"
  convocales=0
  for letra in cadena:
    for vocal in vocales:
      if letra==vocal:
        convocales +=1
  return convocales
cadena=input("Ingrese una cadena: ")
contador=contadorvucal(cadena)
print("El nombre tiene",contador,"vocales")
```

#****Ejercicio de cadena N2****

Crea una funcion que reciba una cadena y devuelva la cadena invertida

```python
def invertir(cadena):
  invertida=""
  for x in range(len(cadena)-1,-1,-1):
    invertida += cadena[x]
  return invertida
cadena=input("Ingrese una cadena: ")
invertida=invertir(cadena)
print("La cadena invertida es:",invertida)
```

```python
def invertir(cadena):
  invertida=cadena[::-1]
  return invertida
cadena=input("Ingrese una cadena: ")
invertida=invertir(cadena)
print("La cadena invertida es:",invertida)
```

#****Ejercicio de cadena N4****

Crea una funcion que elimine todos los espacios en blando de una cadena

```python
def quitarespa(cadena):
  sinespacios=""
  for letra in cadena:
    if letra != " ":
      sinespacios += letra
  return sinespacios
cadena=input("Ingrese una cadena: ")
sinespacios=quitarespa(cadena)
print("La cadena sin espacios es:",sinespacios)
```

```python
def quitarespa(cadena):
  return cadena.replace(" ","")
cadena=input("Ingrese una cadena: ")
sinespacios=quitarespa(cadena)
print("La cadena sin espacios es:",sinespacios)
```

#****Ejecicio de cadena N5****

Escribe una funcion que verifique si una cadena es un palindromo(se lee igual al derecho y al reves)

```python
def verificador(cadena):
  invertida=cadena[::-1]
  return invertida
cadena=input("Ingrese una cadena: ")
invertida=verificador(cadena)
if invertida==cadena:
  print("La cadena invertida es",invertida,"y si es un palindromo")
else:
  print("La cadena invertida es",invertida,"y no es un palindromo")
```

#****Ejercicio de cadena N6****

Escribe un programa que reemplace todas las vocales por asteriscos(\*)en una cadena dada

```python
cadena=input("Ingrese una cadena: ")
vocales="aeiouAEIOUáéíóúÁÉÍÓÚ"
for vocal in vocales:
  cadena=cadena.replace(vocal,"*")
print("La cadena con las vocales reemplazadas es:",cadena)
```

#****Ejercicio de cadena N7****

Escribe una funcion que reciba una letra y una cadena, y cuente cuantas veses aparece esa letra

```python
def localiazar(letra,cadena):
  contador=0
  for x in cadena:
    if x==letra:
      contador +=1
  return contador
letra=input("Ingrese una letra: ")
cadena=input("Ingrese una cadena: ")
contador=localiazar(letra,cadena)
print("La letra",letra,"aparece",contador,"veces en la cadena")
```

#****Ejercicio de cadena N9****

Escribe una funcion que extraiga el dominio de una direccion de correo electronico

```python
def extrador(correo):
  dominio=correo.split("@")[1]
  return dominio
correo=input("Ingrese una direccion de correo electronico: ")
dominio=extrador(correo)
print("El dominio de la direccion de correo electronico es:",dominio)
```

#****Ejercicio de cadena N10****

Crea una funcion que compare dos cadenas y determine si son iguales sin importar mayusculas o minusculas

```python
def verificador(cadena1,cadena2):
  cadena1=cadena1.lower()
  cadena2=cadena2.lower()
  if cadena1==cadena2:
    print("Las cadenas son iguales")
  else:
    print("Las cadenas son diferentes")
cadena1=input("Ingrese una cadena: ")
cadena2=input("Ingrese otra cadena: ")
verificador(cadena1,cadena2)
```
