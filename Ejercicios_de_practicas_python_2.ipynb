#****Ejercicio N2****

Determinar si un numero es capicua (palindromo numerico)

```python
num=int(input("Ingrese un numero: "))
capicua=num
palin=0
while num>0:
  dig=num%10
  palin=palin*10+dig
  num=num//10
if capicua==palin:
  print("El numero",capicua,"si es capicua")
else:
  print(f"El numero {capicua} no es capicua")
```

#****Ejercicio N3****

Calcular el area de un circulo para radios del 1 al N

```python
num=int(input("Ingrese un numero: "))
pi=float(3.14)
for x in range(1,num+1):
  area=pi*x**2
  print(f"El area del circulo de radio {x} es {area}")
```

#****Ejercicio N4****

Calcular la hipotenusa de triangulos con catetos enteros(1 a N)

```python
num=int(input("Ingrese un numero: "))
for x in range(1,num+1):
  hip=(x**2+x**2)**0.5
  print(f"La hipotenusa del triangulo de catetos {x} y {x} es {hip}")
```

#****Ejercicio N5****

Encontrar todos los numeros armstrong de 3 digitos

```python
num=100
while 100<=num<=999:
  dig1=num//100
  dig2=(num%100)//10
  dig3=num%10
  if num==dig1**3+dig2**3+dig3**3:
    print("El numero",num,"es un numero de armstrong")
  num+=1
```

#****Ejercicio N6****

Calcular el coeficiente binomial (n sobre K) usando multiplicaciones y diviciones

image.png

```python
n=int(input("Ingrese el valor de n: "))
k=int(input("Ingrese el valor de k: "))
factn=1
for x in range(1,n+1):
  factn*=x
factk=1
for x in range(1,k+1):
  factk*=x
factnk=1
for x in range(1,n-k+1):
  factnk*=x
coeficiente=factn/(factk*factnk)
print("El coeficiente binomial de n sobre k es",coeficiente)
```

```python
def factorial(num):
  fac=1
  for x in range(1,num+1):
    fac*=x
  return fac
n=int(input("Ingrese el valor de n: "))
facN=factorial(n)
k=int(input("Ingrese el valor de k: "))
facK=factorial(k)
factnk=factorial(n-k)
coeficiente=facN/(facK*factnk)
print("El coeficiente binomial de n sobre k es",coeficiente)
```
