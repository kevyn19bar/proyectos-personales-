#****Ejercicio N1****
Una empresa quiere registrar los nombres de sus empleados por departamento. Crea
un diccionario donde las claves sean los nombres de los departamentos y los valores,
una lista con los nombres de los empleados de cada uno. Permite consultar los
empleados de un departamento ingresado por el usuario

```python
empresa={}
for x in range(int(input("How many departments are there in the company?: "))):
  departamento=input("Enter the department: ")
  empleados=[]
  for i in range(int(input(f"How many employees are there in {departamento}?: "))):
    empleados.append(input("Enter the employee's name: "))
  empresa[departamento]=empleados
buscar=input("Enter the department you want to search for: ")
if buscar in empresa:
  for x in empresa[buscar]:
    print(x,end=",")
else:
   print("The department does not exist")
```

#****Ejercicio N2****

donde cada clave sea el nombre del producto y el valor sea otro diccionario con la
cantidad disponible y el precio. Permite consultar el total del inventario valorizado

```python
productos={}
total_inventario = 0
for i in range(int(input("How many products are you going to register?: "))):
  nombre = input(f"Enter the product name #{i+1}: ")
  cantidad= int(input("enter the quantity of the product: "))
  precio= float(input(" enter the price of the product: "))
  valor=cantidad*precio
  productos[nombre] = {"amount": cantidad,"price": precio, "total_value": valor}
  total_inventario += valor
for nombre, info in productos.items():
    print(f"Product: {nombre}")
    print(f"  - Amount: {info['amount']}")
    print(f"  - Unit price: ${info['price']}")
    print(f"  - Stock value: ${info['total_value']}\n")
if input("Do you want to see the total valued inventory? (yes/no): ").lower() == "yes":
  print(f"The total value of the inventory is: ${total_inventario:,.2f}")
```

#****Ejercicio N4****
Registra las ventas realizadas por cada vendedor durante una semana. Usa un
diccionario donde la clave sea el nombre del vendedor y el valor sea una lista con el
total vendido cada día. Calcula el total semanal por vendedor y el mejor vendedor

```python
venta_realizada={}
while True:
  key=input("enter the seller's name: ")
  venta_realizada[key]=[]
  if input("Do you want to add another seller? (yes/no): ").lower()=="no":
    break
for i in range(1,8):
  for vendedor in venta_realizada:
      valor = float(input(f"Enter the total sold by {vendedor} in the day {i}: "))
      venta_realizada[vendedor].append(valor)
total_semanal = {}
for vendedor, ventas_diarias in venta_realizada.items():
  total_semanal[vendedor] = sum(ventas_diarias)
print("\n--- Weekly Sales Summary ---")
best_vendedor=None
max_ventas= -1
for vendedor, total in total_semanal.items():
  print(f"{vendedor}: ${total:,.2f}")
  if total > max_ventas:
    max_ventas = total
    best_vendedor = vendedor
print(f"\nThe best seller of the week is: {best_vendedor} with a total of ${max_ventas:,.2f}")
```

#****Ejercicio N5****
Una empresa ofrece varios servicios. Crea un diccionario donde la clave sea el nombre
del servicio y el valor una descripción y el precio. Permite listar todos los servicios y
buscar por nombre para ver su información.


```python
servicios = {}
def menu():
    print("\n--- COMPANY SERVICES MENU ---\n1. Add a service\n2. List all services\n3. Search for a service by name\n4. Exit")
    print("-"*40)
    return input("Select an option: ")
def agregar_servicio():
    print("\n--- ADD NEW SERVICE ---")
    nombre = input("Enter the name of the new service: ").strip().title()
    if nombre in servicios:
        print(f"\n Error: The service '{nombre}' already exists.")
        return
    descripcion = input("Enter the service description: ").strip()
    precio = float(input("Enter the price of the service (e.g. 350.50): "))
    servicios[nombre] = {"description": descripcion,"price": precio}
    print(f"\n Service '{nombre}' successfully added!")
def listar_servicios():
    print("\n--- COMPLETE LIST OF SERVICES ---")
    if not servicios:
        print("No services have been added yet.")
        return
    for nombre, detalles in servicios.items():
        print(f"\n Service: {nombre}")
        print(f"  description: {detalles['description']}")
        print(f"   price: ${detalles['price']:.2f}")
    print("-"*40)
def buscar_servicio():
    print("\n--- SERVICE SEARCH ---")
    nombre_buscado = input("Enter the name of the service you want to search for: ").strip()
    servicio_encontrado = None
    clave_encontrada = ""
    for nombre_servicio in servicios:
        if nombre_servicio.lower() == nombre_buscado.lower():
            servicio_encontrado = servicios[nombre_servicio]
            clave_encontrada = nombre_servicio
            break
    if servicio_encontrado:
        print(f"\n Service information '{clave_encontrada}':")
        print(f"  description: {servicio_encontrado['description']}")
        print(f"  price: ${servicio_encontrado['price']:.2f}")
    else:
        print(f"\n Sorry, the service '{nombre_buscado}' was not found.")
while True:
  match menu():
    case "1":
      agregar_servicio()
    case "2":
      listar_servicios()
    case "3":
      buscar_servicio()
    case "4":
      break
    case _:
      print("\n Invalid option. Please choose an option from 1 to 4..")
```

#****Ejercicio N7****
Crea un diccionario que almacene la evaluación de desempeño de cada empleado (de
0 a 100). Muestra los empleados con desempeño mayor a 85 y calcula el promedio
general.

```python
def menu():
  print("\nMENU\n 1. Add employee and their performance\n 2. Show performances\n 3. Calculate the average\n 4. Exit")
  return input("Choose an option: ")
def add_employee(evaluations, employee, performance):
  evaluations[employee] = performance
def show_performance(evaluations):
  print("\nEmployees with Performance Higher than 85")
  print("*" * 40)
  for employee, performance in evaluations.items():
    if performance > 85:
      print(f"- {employee}: {performance}")
def calculate_average(evaluations):
  total_performance = sum(evaluations.values())
  total_employees = len(evaluations)
  print("\nGeneral Performance Average")
  print("*" * 40)
  print(f"The general performance average of all employees is: {total_performance / total_employees:.2f}")
evaluations = {}
while True:
  match menu():
    case "1":
      add_employee(evaluations, input("\nEnter the employee's name: "), int(input("Enter the employee's performance: ")))
    case "2":
      show_performance(evaluations)
    case "3":
      calculate_average(evaluations)
    case "4":
      break
    case _:
      print("Invalid option")
```

#****Ejercicio N8****
Una empresa hizo una encuesta a sus clientes donde calificaron del 1 al 5 varios
aspectos: atención, calidad, puntualidad. Usa un diccionario por cliente (clave: correo
electrónico) con otro diccionario como valor para registrar sus calificaciones. Calcula
promedios por aspecto.

```python
survey = {}
tudata=("attention","quality","punctuality")
while True:
  email= input("Enter respondent's email: ")
  survey[email]={}
  for data in tudata:
    survey[email][data]=int(input(f"Rate company's {data} attention from 1 to 5: "))
  if input("\nDo you want to add another respondent (yes/no): ").lower()=="no":
    break
total_service_attention = 0
total_quality = 0
total_punctuality = 0
number_of_respondents = len(survey)
for email, ratings in survey.items():
    total_service_attention += ratings["service_attention"]
    total_quality += ratings["quality"]
    total_punctuality += ratings["punctuality"]
print("\nSatisfaction Survey Results")
print("-" * 40)
print(f"Total surveys received: {number_of_respondents}")
print("\nAverage Rating per Aspect")
print("-" * 40)
print(f"Customer Service Attention: {total_service_attention / number_of_respondents:.2f}")
print(f"Product/Service Quality: {total_quality / number_of_respondents:.2f}")
print(f"Delivery Punctuality: {total_punctuality / number_of_respondents:.2f}")
```

#****Ejercicio N9****
Lleva un historial de pedidos por cliente. Usa un diccionario donde la clave sea el
nombre del cliente y el valor una lista de pedidos, cada uno representado por un
subdiccionario con producto, cantidad y valor total. Permite consultar el historial de un
cliente.

```python
def order_history(history, name, product, quantity, total_value):
    if name not in history:
        history[name] = []
    sub_dictionary = {"product": product, "quantity": quantity, "total_value": total_value}
    history[name].append(sub_dictionary)
    return input(f"Do you want to add another product for {name} (yes/no): ")
def view_history(client_name):
    if client_name in history:
        print(f"Order history for {client_name}")
        orders = history[client_name]
        for i, order in enumerate(orders, 1):
            print(f"   Order N°{i}:")
            print(f"     Product: {order['product']}")
            print(f"     Quantity: {order['quantity']}")
            print(f"     Total Value: ${order['total_value']:.2f}")
    else:
        print(f"No order history found for {client_name}.")
    print("*" * 40)
history = {}
while True:
    client_name = input("Enter client's name: ")
    while True:
        if order_history(history, client_name,
                         input("Enter the product: "),
                         int(input("Enter the quantity of the product: ")),
                         int(input("Enter the total value of the product: "))).lower() == "no":
            break
    if input("Do you want to add another client (yes/no): ").lower() == "no":
        break
view_history(input("Enter the name of the client to view their history: "))
```

```python
def menu():
    print("\n" + "*" * 40)
    print("\t\t MENU")
    print("*" * 40)
    print(" 1. Register Product")
    print(" 2. Generate Bill")
    print(" 3. Exit")
    print("*" * 40)
    return input("Select an option: ")

def percentage(vat):
  match (int(vat)):
    case 1:
      return 0
    case 2:
      return 0.05
    case 3:
      return 0.19

def register_product():
    print("\n--- Register New Product ---")
    while True:
        product_id =  int(input("Enter the Product Code (numbers only): "))
        if product_id in products:
            print("This product code already exists. Please enter a different one.")
        else:
            break
    products[product_id] = {}
    for field in PRODUCT_FIELDS:
        if field == "Count":
            products[product_id][field] = int(input(f"Enter the Product's {field}: "))
        elif field == "Unit Price":
            products[product_id][field] = float(input(f"Enter the Product's {field}: "))
        elif "VAT Type" in field:
            while True:
                vat_type = input(f"Enter the {field}: ")
                if vat_type in ["1", "2", "3"]:
                    (vat_type)
                    products[product_id]["VAT Type"] = vat_type
                    break
                else:
                    print("Invalid VAT type. Please enter 1, 2, or 3.")
        else:
            products[product_id][field] = input(f"Enter the Product's {field}: ")
    print(f"Product '{products[product_id]['Name']}' registered successfully.")

def generate_bill():
    if not products:
        print("\nNo products are registered. Please register products before generating a bill.")
        return
    print("\n--- Generate New Bill ---")
    bill_number = max(bills.keys()) + 1 if bills else 1000
    bills[bill_number] = {}
    for field in CLIENT_FIELDS:
        bills[bill_number][field] = input(f"Enter Client's {field}: ")
    bills[bill_number]["products"] = {}
    total_bill = 0
    while True:
        while True:
            code = int(input("Enter the code of the product to purchase: "))
            if code in products:
                break
            else:
                print("Product code does not exist. Try again.")
        product_info = products[code]
        product_name = product_info["Name"]
        available_Count = product_info["Count"]
        unit_price = product_info["Unit Price"]
        percent=percentage(product_info["VAT Type"])
        print(f"Product: {product_name} | Available: {available_Count} | Price: ${unit_price:.2f} ")
        while True:
            Count_to_buy =  int(input(f"Enter the Count of '{product_name}' to buy: "))
            if Count_to_buy == 0:
                 print("Count must be greater than zero.")
            elif Count_to_buy > available_Count:
                print(f"Not enough stock. Only {available_Count} units of '{product_name}' are available.")
            else:
                break
        products[code]["Count"] -= Count_to_buy
        bills[bill_number]["products"][code] = {"Name": product_name,"Count": Count_to_buy,"Unit Price": unit_price}
        item_total = Count_to_buy * unit_price
        total_bill += item_total
        print(f"'{product_name}' added to the bill. Remaining stock: {products[code]['Count']}")
        if input("Add another product to the purchase? (Yes: press 1 | No: any other key): ") != "1":
            break
    VAT_total_bill=total_bill*percent
    complete_invoice=total_bill+VAT_total_bill
    bills[bill_number]["Total"] = complete_invoice
    print(f"\n--- Bill No. {bill_number} generated successfully. Total: ${complete_invoice:.2f} ---")

def print_products():
    if not products:
        print("\nNo products registered.")
        return
    print("\n--- Registered Products ---")
    for product_id, details in products.items():
        print(f"Code: {product_id}")
        for key, value in details.items():
            print(f"  {key}: {value}")
        print("-" * 20)

def print_bills():
    if not bills:
        print("\nNo bills generated.")
        return
    print("\n--- Generated Bills ---")
    for bill_id, details in bills.items():
        print(f"Bill No. {bill_id}")
        print("  Client Information:")
        for key, value in details.items():
            if key not in ["products", "Total"]:
                print(f"    {key}: {value}")
        print("  Purchased Products:")
        for code, prod_details in details["products"].items():
            print(f"    Code: {code}, Name: {prod_details['Name']}, Count: {prod_details['Count']}, Unit Price: ${prod_details['Unit Price']:.2f}")
        print(f"  Bill Total: ${details['Total']:.2f}")
        print("-" * 30)
PRODUCT_FIELDS = ("Name", "Count", "Unit Price", "VAT Type (1. No VAT | 2. 5% | 3. 19%)")
CLIENT_FIELDS = ("Identification", "Name", "Address", "Phone")
products = {}
bills = {}
while True:
    match (menu()):
      case '1':
        register_product()
      case '2':
        generate_bill()
      case '3':
        print("Exiting the program. Goodbye!")
        break
      case '4':
        print_bills()
      case '5':
        print_products()
      case _:
        print("Invalid option, please try again!")
```
